## ğŸ“¦ ë°ì´í„° ì¤€ë¹„ (ì‹¤í–‰ë§Œ í•˜ì„¸ìš”)

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

iowa_file_path = '../data/home-data-for-ml-course/train.csv'
home_data = pd.read_csv(iowa_file_path)

y = home_data.SalePrice
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF',
            'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[features]
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
print('ë°ì´í„° ì¤€ë¹„ ì™„ë£Œ!')
```

## ğŸ“Š ë¬¸ì œ 1: ê²°ì • íŠ¸ë¦¬ MAE í™•ì¸ (ë¹„êµìš©)

ê²°ì • íŠ¸ë¦¬ ëª¨ë¸ì˜ MAEë¥¼ êµ¬í•´ë‘ì„¸ìš” (ëœë¤ í¬ë ˆìŠ¤íŠ¸ì™€ ë¹„êµìš©).

```python
from sklearn.tree import DecisionTreeRegressor

# ê²°ì • íŠ¸ë¦¬ (ë¹„êµ ê¸°ì¤€)
dt_model = DecisionTreeRegressor(random_state=1)
dt_model.fit(train_X, train_y)
dt_preds = dt_model.predict(val_X)
dt_mae = mean_absolute_error(val_y, dt_preds)

print(f'ê²°ì • íŠ¸ë¦¬ MAE: ${dt_mae:,.0f}')
```
