```python
# ì •ë‹µ ì½”ë“œ
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

home_data = pd.read_csv('../data/home-data-for-ml-course/train.csv')

y = home_data.SalePrice

iowa_features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF',
                 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
X = home_data[iowa_features]

iowa_model = DecisionTreeRegressor(random_state=1)
iowa_model.fit(X, y)

predictions = iowa_model.predict(X.head())
print('ì˜ˆì¸¡ ê°€ê²©:', predictions)
print('ì‹¤ì œ ê°€ê²©:', y.head().values)
```
